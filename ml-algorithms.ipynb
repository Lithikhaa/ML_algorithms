```python
#counter - to count the frequency of each index in the outlier_indices list.  
from collections import Counter
import numpy as np

def detect_outliers(df, features):
    outlier_indices = []
    for c in features:
        # 1st quartile
        Q1 = np.percentile(df[c], 25)
        # 3rd quartile
        Q3 = np.percentile(df[c], 75)
        # IQR
        IQR = Q3 - Q1
        # Outlier step
        outlier_step = IQR * 1.5
        # detect outlier and their indices
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # store indices
        outlier_indices.extend(outlier_list_col)
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = [i for i, v in outlier_indices.items() if v > 1]
    return multiple_outliers
```

```python
# Doctor A's evaluations:

# 10 patients rated as "mild"
# 15 patients rated as "moderate"
# 5 patients rated as "severe"
# Doctor B's evaluations:

# 8 patients rated as "mild"
# 12 patients rated as "moderate"
# 10 patients rated as "severe"
# To calculate Cohen's kappa score, we need to create a contingency table that shows the frequency of agreement and disagreement between the two raters.

# Contingency Table:

# arduino
# Copy code
#             Doctor B
#             "mild"  "moderate"  "severe"
# Doctor A
# "mild"       8         2           0
# "moderate"   0         10          5
# "severe"     0         0           5
# The diagonal cells in the contingency table represent the cases where both raters agreed (i.e., "mild" with "mild", "moderate" with "moderate", and "severe" with "severe"). The off-diagonal cells represent the cases where they disagreed.

# Next, we calculate the observed agreement (Po), which is the proportion of cases where both raters agreed:

# scss
# Copy code
# Po = (number of cases where both raters agreed) / (total number of cases)
# Po = (8 + 10 + 5) / (10 + 15 + 5) = 0.6
# Then, we calculate the expected agreement by chance (Pe), which is the probability of agreement by chance:

# scss
# Copy code
# Pe = (sum of (row total * column total)) / (total number of cases)^2
# Pe = ((10 + 0 + 0) * (0 + 10 + 5) + (0 + 15 + 0) * (8 + 0 + 0) + (0 + 0 + 5) * (0 + 0 + 5)) / (10 + 15 + 5)^2
# Pe = 0.36
# Finally, we can calculate Cohen's kappa score using the formula:

# scss
# Copy code
# Cohen's kappa score = (Po - Pe) / (1 - Pe)
# Cohen's kappa score = (0.6 - 0.36) / (1 - 0.36)
# Cohen's kappa score = 0.4
# A Cohen's kappa score of 0.4 indicates moderate agreement between the two raters. Higher values of Cohen's kappa score (closer to 1) indicate higher levels of agreement, while lower values (closer to -1) indicate lower levels of agreement. A score of 0 indicates agreement that is no better than chance.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import f1_score

X = preprocessor.fit_transform(df.iloc[:, :-1])  # All columns except for the last three columns
Y = df.iloc[:, -1] 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) # provide a value for test_size

accuracies = {}
kappaScores = {}
f1scores = {}
```
